```python
import pandas as pd
```

```python
df_data = pd.read_parquet('green_tripdata_2025-11.parquet')
```

```python
print(df_data.head())
print(df_data.dtypes)
print(df_data.shape)
```

```python
dtype_zone = {
    "LocationID": "int",
    "Borough": "string",
    "Zone": "string",
    "service_zone": "string"
}
df_zone = pd.read_csv(
        "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv",
        dtype=dtype_zone
    )
```

```python
print(df_zone.head())
print(df_zone.dtypes)
print(df_zone.shape)
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')
```

```python
df_data.to_sql(name='green_taxi_data', con=engine, if_exists='replace')
```

```python
df_zone.to_sql(name='taxi_zone_lookup', con=engine, if_exists='replace')
```
